**So sánh biến định lượng hai nhóm độc lập**

Trước khi so sánh biến định lượng của hai nhóm độc lập cần kiểm tra giả định thống kê: (1) Kiểm tra có phân phối chuẩn hay không? (2) Kiểm tra phương sai hai nhóm khác nhau có ý nghĩa không?

Kiểm tra phân phối chuẩn có thể dùng Shapiro test (p>0,05) hoặc dùng phân tán đồ (dạng hình chuông) đều được chấp nhận, giá trị ngang nhau.

`shapiro.test(dat$vas2)` #Kiểm định phân phối chuẩn

`hist(dat$vas2)` #Phân tán đồ

`var.test(dat$vas2~dat$grp)` #So sánh phương sai

`t.test(dat$vas2~dat$grp)` #So sánh độc lập, phương sai khác nhau có ý nghĩa

`t.test(dat$vas2~dat$grp,var.equal=TRUE)` #So sánh độc lập, phương sai không khác nhau có ý nghĩa

`wilcox.test(dat$vas2~dat$grp)` #So sánh độc lập, không phân phối bình thường


```r
shapiro.test(dat$vas2)
```

Điểm đau theo VAS ở thời điểm can thiệp 2 tuần không có phân phối chuẩn (p<0,05)

```r
hist(dat$vas2)
```

Điểm đau theo VAS ở thời điểm can thiệp 2 tuần không có phân phối chuẩn (phân tán đồ không có dạng hình chuông).

```r
var.test(dat$vas2~dat$grp)
```

Phương sai điểm VAS sau 2 tuần can thiệp của hai nhóm chứng và nhĩ châm khác biệt nhau có ý nghĩa thống kê (p<0,05).

```r
wilcox.test(dat$vas2~dat$grp)
```

Điểm VAS sau 2 tuần can thiệp của nhóm nhĩ châm thấp hơn nhóm chứng có ý nghĩa thống kê (p<0,05).

**Hồi quy tuyến tính**

Tiến hành qua 5 bước: (1) Xây dựng mô hình hồi quy; (2) Kiểm tra giả định của mô hình; (3) Lựa chọn mô hình tối ưu; (4) Đánh giá tầm quan trọng của các biến độc lập trong mô hình; (5) Triển khai mô hình bằng nomogram

(1) Xây dựng mô hình

Dùng lệnh lm trong gói cơ bản của R

`m = lm(dat$vas2~dat$sex+dat$vas0+dat$age, data=dat)`

Hiển thị kết quả bằng lệnh summary

`summary(m)`

```r
m = lm(vas2~sex+vas0+age+grp, data=dat)
summary(m)
```

***Đọc kết quả***: Điểm VAS trước can thiệp (vas0) và nhóm điều trị ảnh hưởng có ý nghĩa đến điểm VAS kết thúc can thiệp (p<0,05). Các yếu tố giới tính (sex) và tuổi (age) không ảnh hưởng có ý nghĩa đến điểm VAS sau can thiệp (p>0,05).

Adjusted R-squared (R bình phương hiệu chỉnh): toàn bộ mô hình giải thích được 55,27% sự thay đổi của điểm VAS lúc kết thúc can thiệp

**(3) Lựa chọn mô hình tối ưu**

Cài đặt và gọi gói phần mềm BMA

`library(BMA)`

Xác định biến phụ thuộc (yvar) và biến độc lập (xvar)

`yvar = dat[, ("vas2")]`

`xvars =dat[, c("sex", "vas0","age","grp")]`

Dùng lệnh bicreg để tính toán mô hình

Khai báo kết quả bằng lệnh summary

```r
library(BMA)
yvar = dat[, ("vas2")]
xvars =dat[, c("sex", "vas0","age","grp")]
bma = bicreg(xvars, yvar, strict=FALSE, OR=20)
summary(bma)
```

Mô hình tiên lượng điểm VAS sau 2 tuần can thiệp tốt nhất theo BMA đề xuất là mô hình gồm 2 biến độc lập là nhóm can thiệp và điểm VAS trước can thiệp. Mô hình với hai biến này có thể giải thích 55,2% sự thay đổi của điểm VAS sau 2 tuần can thiệp (R2 = 0,552). Xác suất xuất hiện mô hình tiên lượng này khi lặp lại nghiên cứu nhiều lần là 74,3%.
